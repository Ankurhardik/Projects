```python
dt['Car_Vol']=dt['width_mm']*dt['length_mm']*dt['height_mm']
```

```python
dt.drop(columns=['width_mm','length_mm','height_mm','source_url','number_of_cells','model'],inplace=True)
```

```python
non_numeric = dt[~dt['cargo_volume_l'].apply(lambda x: pd.to_numeric(x, errors='coerce')).notnull()]
print(f"Non-numeric values count: {len(non_numeric)}")
```

```python
dt['cargo_volume_l'] = pd.to_numeric(dt['cargo_volume_l'], errors='coerce')
```

```python
dt.dropna(inplace=True)
```

```python
dt['fast_charge_port'].value_counts()
```

```python
dt['battery_type'].value_counts()
```

```python
dt.drop(columns=['fast_charge_port','battery_type'],inplace=True)
```

```python
brand_counts = dt['brand'].value_counts()
rare_brands = brand_counts[brand_counts < 20].index
dt['brand'] = dt['brand'].apply(lambda x: 'Other' if x in rare_brands else x)
```

```python
dt_numeric = dt.select_dtypes(include='number')
dt_numeric.corr()
```

```python
dt.drop(columns=['Car_Vol','efficiency_wh_per_km','towing_capacity_kg','seats','cargo_volume_l'],inplace=True)
```

```python
dt['segment'].value_counts()
```

```python
dt['segment'] = dt['segment'].str.split('-').str[1].str.strip()
```

```python
dt['drivetrain'].value_counts()
```

```python
dt['car_body_type'].value_counts()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(dt.drop(columns=['range_km']),dt['range_km'],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
categorical_features = ['brand', 'segment','drivetrain','car_body_type']
numeric_features = ['top_speed_kmh','battery_capacity_kWh','torque_nm','acceleration_0_100_s','fast_charging_power_kw_dc']
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])
```

```python
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('model', XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42))
])
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
pipeline
```

```python
pipeline.fit(X_train, Y_train)
Y_pred = pipeline.predict(X_test)
```

```python
Y_pred=Y_pred.astype(int)
```

```python
Y_pred
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Y_test, y=Y_pred, alpha=0.6, color='teal')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')
plt.xlabel('Actual Range (km)')
plt.ylabel('Predicted Range (km)')
plt.title('Predicted vs Actual Range')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y_test, Y_pred)
print("R² Score:", r2)
```
