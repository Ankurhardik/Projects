```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dt=pd.read_csv('train.csv')
df=pd.read_csv('test.csv')
test=df.copy()
```

```python
dt.head()
```

```python
dt['Age'] = dt['Age'].fillna(dt['Age'].median())
df['Age'] = df['Age'].fillna(dt['Age'].median())
```

```python
df
```

```python
dt.drop(columns=['PassengerId', 'Name', 'Cabin','Ticket'], inplace=True, errors='ignore')
df.drop(columns=['PassengerId', 'Name', 'Cabin','Ticket'], inplace=True, errors='ignore')
```

```python
dt.head()
```

```python
dt['Embarked'] = dt['Embarked'].fillna(dt['Embarked'].mode()[0])
```

```python
dt.info()
```

```python
df['Fare'] = df['Fare'].fillna(dt['Fare'].median())
```

```python
df.info()
```

```python
dt.describe()
```

```python
cat=dt[['Sex','Embarked']]
cat1=df[['Sex','Embarked']]
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)
cat=ohe.fit_transform(cat)
cat1=ohe.transform(cat1)
```

```python
cat_df = pd.DataFrame(cat, columns=ohe.get_feature_names_out(['Sex', 'Embarked']), index=dt.index)
cat1_df = pd.DataFrame(cat1, columns=ohe.get_feature_names_out(['Sex', 'Embarked']), index=df.index)
dt = pd.concat([dt, cat_df], axis=1)
df = pd.concat([df, cat1_df], axis=1)
```

```python
dt.head()
```

```python
df
```

```python
dt.drop(columns=['Sex','Embarked'],inplace=True,errors='ignore')
df.drop(columns=['Sex','Embarked'],inplace=True,errors='ignore')
```

```python
dt['Family']=dt['SibSp']+dt['Parch']
df['Family']=df['SibSp']+df['Parch']
```

```python
dt.drop(columns=['SibSp','Parch'],inplace=True,errors='ignore')
df.drop(columns=['SibSp','Parch'],inplace=True,errors='ignore')
```

```python
df
```

```python
Y=dt.iloc[:,0]
X=dt.iloc[:,1:]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
SS=StandardScaler()
X_train=SS.fit_transform(X_train)
X_test=SS.transform(X_test)
df=SS.transform(df)
```

```python
df
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(4,activation='relu',input_dim=7))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(X_train,Y_train,epochs=100,validation_split=0.2)
```

```python
Y_log=model.predict(X_test) 
Y_pred=np.where(Y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(Y_test,Y_pred)
acc
```

```python
from sklearn.linear_model import LogisticRegression
Lreg=LogisticRegression(max_iter=800)
Lreg.fit(X_train,Y_train)
```

```python
Y_pred1=Lreg.predict(X_test)
ans=Lreg.predict(df)
```

```python
accuracy=accuracy_score(Y_test,Y_pred1)
```

```python
final_data = {'PassengerId': test.PassengerId, 'Survived': ans}
submission = pd.DataFrame(data=final_data)
submission.to_csv('submission.csv', index =False)
```
